```python
import pandas as pd
```

```python
# take the original datasets for ratings with 3000 book id
df_description = pd.read_csv('../datasets/book_description_dataset/description.csv')
df_description.head()
```

```python
df_description.drop(columns = ['book_name'], inplace = True)
```

```python
df_description
```

```python
#take the original dataset for book description with 3000 rows
df_ratings = pd.read_csv('../datasets/book_rating_dataset/ratings.csv')
df_ratings.head()
```

```python
# dropping unnamed columns
df_ratings.drop(columns = ['Unnamed: 0','Unnamed: 0.1', 'user_id', 'rating'], inplace = True)
```

```python
# only take book ids upto 3000 to match the description dataframe
df_ratings = df_ratings[df_ratings['book_id'] <3000]
df_ratings.drop_duplicates(keep='first', inplace = True)
```

```python
hybrid_df = pd.merge(df_description,df_ratings, on=['book_id'], how='left')
```
